```python
# RETROELEMENTS

with open("./epigenomic_files/map_epigenomes_on_transposons.sh", 'w') as script:
# --- Start of Bash Script ---
    
    # Add a shebang and a command to make the script exit on error
    script.write("#!/bin/bash\n")
    script.write("set -e\n\n")

    # Define ANSI color codes for pretty printing in the bash script
    script.write("# Define some colors for logging\n")
    script.write("GREEN='\\033[0;32m'\n")
    script.write("BLUE='\\033[0;34m'\n")
    script.write("NC='\\033[0m' # No Color\n\n")

    # Add a starting message
    script.write("echo -e \"${GREEN}Starting epigenome mapping process...${NC}\\n\"\n")

    # Loop through each file and write the commands and logs
    for i, file in enumerate(files_final):
        bedgraph_name = file.replace('bw', 'bedGraph')
        mapped_name = file.replace('bw', 'mapped_on_repeat_masker.bedGraph')

        # Main processing block for each file
        script.write(f"# --- Processing file {i+1}/{len(files_final)} ---\n")
        script.write(f"echo -e \"${{BLUE}}Processing: {file}${{NC}}\"\n")

        # Step 1: Convert bigWig to bedGraph
        script.write("echo '  Step 1/3: Converting bigWig to bedGraph and sorting...'\n")
        script.write(f"../bigWigToBedGraph {file} stdout | sort -k1,1 -k2,2n > {bedgraph_name}\n")

        # Step 2: Map with bedtools
        script.write("echo '  Step 2/3: Mapping features with bedtools...'\n")
        script.write(f"bedtools map -a ../T2T_repeat_masker_processed_sorted.bed -b {bedgraph_name} -c 4 -o mean > {mapped_name}\n")

        # Step 3: Clean up intermediate file
        script.write("echo '  Step 3/3: Cleaning up intermediate files...'\n")
        script.write(f"rm {bedgraph_name}\n")
        script.write(f"rm {file}\n")
        
        # Success message for the file
        script.write(f"echo -e \"${{GREEN}}Successfully mapped {file} to {mapped_name}${{NC}}\\n\"\n")

    # Add a final completion message
    script.write("echo -e \"${GREEN}All files processed successfully!${NC}\"\n")
    # The 'with open(...) as script:' block handles closing the file automatically
```
